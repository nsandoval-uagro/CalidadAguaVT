```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.feature_selection import SelectKBest, f_regression
import warnings
warnings.filterwarnings('ignore') # Para limpiar la consola de advertencias

# 1. Cargar datos
data = pd.read_csv('/content/drive/MyDrive/Datos CA VT/BD Completa.csv')

# 2. Definir variables
bandas = ["Banda_2", "Banda_3", "Banda_4", "Banda_5", "Banda_6",
          "Banda_7", "Banda_8", "Banda_8a", "Banda_11", "Banda_12"]

objetivos = ["pH", "Temperatura", "Conductividad", "Eh", "Nitratos",
             "Sulfatos", "Floruros", "Alcalinidad", "Cloruros"]

# ====================================================================
# PASO CLAVE: ESCALADO DE BANDAS PARA CALCULADORA RASTER
# ====================================================================
# Multiplicamos los valores espectrales por 1000 antes de entrenar los modelos.
# Esto obliga al algoritmo a calcular coeficientes adaptados a valores crudos (grandes).
data[bandas] = data[bandas] * 1000

resultados = []

print("--- Evaluación Multivariante: Lineal, Exponencial y Polinómico ---")

for var_y in objetivos:
    # Eliminar nulos
    df_temp = data[bandas + [var_y]].dropna()
    X = df_temp[bandas]
    y = df_temp[var_y]

    print(f"\n{'='*50}")
    print(f"ANALIZANDO PARÁMETRO: {var_y}")
    print(f"{'='*50}")

    # Configurar figura para mostrar los 3 modelos lado a lado
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle(f'Comparativa de Modelos para {var_y}', fontsize=16)

    # ---------------------------------------------------------
    # MODELO 1: REGRESIÓN LINEAL MÚLTIPLE
    # ---------------------------------------------------------
    modelo_lin = LinearRegression()
    modelo_lin.fit(X, y)
    y_pred_lin = modelo_lin.predict(X)

    r2_lin = r2_score(y, y_pred_lin)
    rmse_lin = np.sqrt(mean_squared_error(y, y_pred_lin))

    ecuacion_lin = f"{var_y} = {modelo_lin.intercept_:.4f}"
    for i, b in enumerate(bandas):
        signo = "+" if modelo_lin.coef_[i] >= 0 else "-"
        ecuacion_lin += f" {signo} {abs(modelo_lin.coef_[i]):.6f}*{b}" # Usamos 6 decimales porque los coeficientes serán pequeños

    resultados.append({'Variable': var_y, 'Tipo': 'Lineal', 'R2': r2_lin, 'RMSE': rmse_lin})

    # Gráfico Lineal
    axes[0].scatter(y, y_pred_lin, color='blue', alpha=0.6)
    axes[0].plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2)
    axes[0].set_title(f'Lineal (R² = {r2_lin:.4f})')
    axes[0].set_xlabel('Valor Real')
    axes[0].set_ylabel('Predicción')
    axes[0].grid(True, linestyle='--', alpha=0.5)

    print("\n[1] MODELO LINEAL")
    print(f"R2: {r2_lin:.4f} | RMSE: {rmse_lin:.4f}")
    print(f"Ecuación:\n{ecuacion_lin}")

    # ---------------------------------------------------------
    # MODELO 2: REGRESIÓN EXPONENCIAL MÚLTIPLE (CORREGIDO)
    # ---------------------------------------------------------
    mask_exp = y > 0
    if mask_exp.sum() > 4: # Se requieren al menos 5 datos
        X_exp_completo = X[mask_exp]
        y_exp = y[mask_exp]
        y_log_target = np.log(y_exp)

        # Corrección: Seleccionar solo 3 bandas + Ridge para evitar sobreajuste
        selector_exp = SelectKBest(score_func=f_regression, k=3)
        X_base_exp = selector_exp.fit_transform(X_exp_completo, y_log_target)
        bandas_exp = X_exp_completo.columns[selector_exp.get_support()].tolist()

        modelo_exp = Ridge(alpha=1.0)
        modelo_exp.fit(X_base_exp, y_log_target)

        y_pred_exp = np.exp(modelo_exp.predict(X_base_exp))

        r2_exp = r2_score(y_exp, y_pred_exp)
        rmse_exp = np.sqrt(mean_squared_error(y_exp, y_pred_exp))

        ecuacion_exp = f"{var_y} = exp( {modelo_exp.intercept_:.4f}"
        for i, b in enumerate(bandas_exp):
            signo = "+" if modelo_exp.coef_[i] >= 0 else "-"
            ecuacion_exp += f" {signo} {abs(modelo_exp.coef_[i]):.6f}*{b}"
        ecuacion_exp += " )"

        resultados.append({'Variable': var_y, 'Tipo': 'Exponencial (Ridge)', 'R2': r2_exp, 'RMSE': rmse_exp})

        # Gráfico Exponencial
        axes[1].scatter(y_exp, y_pred_exp, color='green', alpha=0.6)
        axes[1].plot([y_exp.min(), y_exp.max()], [y_exp.min(), y_exp.max()], 'k--', lw=2)
        axes[1].set_title(f'Exponencial (R² = {r2_exp:.4f})')
        axes[1].set_xlabel('Valor Real')
        axes[1].grid(True, linestyle='--', alpha=0.5)

        print("\n[2] MODELO EXPONENCIAL (Ridge 3 variables)")
        print(f"R2: {r2_exp:.4f} | RMSE: {rmse_exp:.4f}")
        print(f"Ecuación:\n{ecuacion_exp}")
    else:
        axes[1].text(0.5, 0.5, 'Datos no válidos\npara Exponencial (y<=0)', ha='center', va='center', fontsize=12)
        print("\n[2] MODELO EXPONENCIAL: Omitido (contiene valores cero o negativos)")

    # ---------------------------------------------------------
    # MODELO 3: REGRESIÓN POLINÓMICA MULTIVARIANTE (Corregido)
    # ---------------------------------------------------------
    # Seleccionar 3 mejores bandas
    selector_poly = SelectKBest(score_func=f_regression, k=3)
    X_poly_base = selector_poly.fit_transform(X, y)
    bandas_poly_ganadoras = X.columns[selector_poly.get_support()].tolist()

    # Polinomio grado 2
    poly = PolynomialFeatures(degree=2, include_bias=False)
    X_poly = poly.fit_transform(X_poly_base)
    nombres_poly = poly.get_feature_names_out(bandas_poly_ganadoras)

    # Ridge para evitar sobreajuste
    modelo_poly = Ridge(alpha=1.0)
    modelo_poly.fit(X_poly, y)
    y_pred_poly = modelo_poly.predict(X_poly)

    r2_poly = r2_score(y, y_pred_poly)
    rmse_poly = np.sqrt(mean_squared_error(y, y_pred_poly))

    coef_poly = modelo_poly.coef_
    ecuacion_poly = f"{var_y} = {modelo_poly.intercept_:.4f}"

    for i, nombre in enumerate(nombres_poly):
        nombre_limpio = nombre.replace(" ", " * ")
        signo = "+" if coef_poly[i] >= 0 else "-"
        # Aumentamos los decimales a 8 para el polinomio debido al escalado
        ecuacion_poly += f" {signo} {abs(coef_poly[i]):.8f}*[{nombre_limpio}]"

    resultados.append({'Variable': var_y, 'Tipo': 'Polinomico (Ridge)', 'R2': r2_poly, 'RMSE': rmse_poly})

    # Gráfico Polinómico
    axes[2].scatter(y, y_pred_poly, color='red', alpha=0.6)
    axes[2].plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2)
    axes[2].set_title(f'Polinómico (R² = {r2_poly:.4f})')
    axes[2].set_xlabel('Valor Real')
    axes[2].grid(True, linestyle='--', alpha=0.5)

    print("\n[3] MODELO POLINÓMICO (Grado 2, Ridge 3 variables)")
    print(f"R2: {r2_poly:.4f} | RMSE: {rmse_poly:.4f}")
    print(f"Ecuación (Truncada para lectura):\n{ecuacion_poly[:300]} ... [Continúa con {len(nombres_poly)} términos]")

    plt.tight_layout()
    plt.show()

# Exportar tabla resumen final
df_resultados = pd.DataFrame(resultados)
print("\n--- TABLA RESUMEN FINAL ---")
print(df_resultados.sort_values(by=['Variable', 'R2'], ascending=[True, False]).to_string(index=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
data = pd.read_csv('/content/drive/MyDrive/Datos CA VT/BD Completa con RGB.csv')

# Seleccionar columnas de interés
variables_in_situ = ['pH', 'Nitratos', 'Sulfatos', 'Eh', 'Conductividad', 'Temperatura', 'Cloruros', 'Floruros', 'Alcalinidad']
bandas_sentinel = ['Banda_1', 'Banda_2', 'Banda_3', 'Banda_4', 'Banda_5', 'Banda_6', 'Banda_7', 'Banda_8', 'Banda_8a', 'Banda_9', 'Banda_11', 'Banda_12']

# Calcular matriz de correlación
correlacion = data[variables_in_situ + bandas_sentinel].corr()

# Filtrar solo las correlaciones entre variables in situ y bandas
correlacion_filtrada = correlacion.loc[bandas_sentinel, variables_in_situ]

# Visualizar con heatmap
plt.figure(figsize=(30, 20))

# Se asigna a la variable 'ax' para poder modificar la barra de color después
ax = sns.heatmap(correlacion_filtrada,
            annot=True,
            cmap='coolwarm',
            vmin=-1, vmax=1,
            fmt=".2f",
            linewidths=0.5,
            annot_kws={"size": 18}) # <-- MODIFICACIÓN: Aumenta el texto de los números dentro de los cuadros

# MODIFICACIÓN: Aumenta el tamaño del título y le da un poco de espacio (pad)
plt.title('', fontsize=30, pad=20)

# MODIFICACIÓN: Aumenta el tamaño del texto de los ejes X e Y
plt.xticks(fontsize=18, rotation=0) # Se añade rotación de 45 grados por si los nombres se empalman
plt.yticks(fontsize=18, rotation=0)

# MODIFICACIÓN: Aumenta el tamaño de los números en la barra de color lateral (colorbar)
cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=18)

# Ajusta el diseño para que nada quede fuera de los márgenes
plt.tight_layout()

plt.show()
```
