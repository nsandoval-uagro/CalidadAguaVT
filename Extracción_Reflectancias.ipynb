Extraemos los valores de reflectancia de los puntos de muestreo

```python
# Función para extraer valores de reflectancia en un punto
def extraer_reflectancia(archivo_banda, punto):
    with rasterio.open(archivo_banda) as src:

        x, y = punto["X"], punto["Y"]
        row, col = src.index(x, y)
        valor = src.read(1, window=((row, row+1), (col, col+1)))
        return valor[0][0]
```

Extraer Valores RGB

```python
# Cargar los puntos de muestreo desde un CSV
puntos_muestreo = pd.read_csv("/content/drive/MyDrive/Datos CA VT/BD valerio trujano.csv")

# Lista de archivos GeoTIFF para las bandas RGB (Sentinel-2)
banda_2 = "/content/drive/MyDrive/Datos CA VT/Banda2.tif"  # Banda 2 - Azul (490 nm)
banda_3 = "/content/drive/MyDrive/Datos CA VT/Banda3.tif"  # Banda 3 - Verde (560 nm)
banda_4 = "/content/drive/MyDrive/Datos CA VT/Banda4.tif"  # Banda 4 - Rojo (665 nm)

# Otras bandas (mantenidas por si las necesitas)
banda_1 = "/content/drive/MyDrive/Datos CA VT/Banda1.tif"
banda_5 = "/content/drive/MyDrive/Datos CA VT/Banda5.tif"
banda_6 = "/content/drive/MyDrive/Datos CA VT/Banda6.tif"
banda_7 = "/content/drive/MyDrive/Datos CA VT/Banda7.tif"
banda_8 = "/content/drive/MyDrive/Datos CA VT/Banda8.tif"
banda_8a = "/content/drive/MyDrive/Datos CA VT/Banda8A.tif"
banda_9 = "/content/drive/MyDrive/Datos CA VT/Banda9.tif"
banda_11 = "/content/drive/MyDrive/Datos CA VT/Banda11.tif"
banda_12 = "/content/drive/MyDrive/Datos CA VT/Banda12.tif"
NDWI = "/content/drive/MyDrive/Datos CA VT/NDWI.tif"
NDCI = "/content/drive/MyDrive/Datos CA VT/NDCI.tif"
RCI = "/content/drive/MyDrive/Datos CA VT/RCI.tif"
CDOM = "/content/drive/MyDrive/Datos CA VT/CDOM.tif"

# Convertir tipos de datos (mantenido de tu código original)
puntos_muestreo['pH'] = puntos_muestreo['pH'].astype(float)
puntos_muestreo['Temperatura'] = puntos_muestreo['Temperatura'].astype(float)
puntos_muestreo['Conductividad'] = puntos_muestreo['Conductividad'].astype(float)
puntos_muestreo['Eh'] = puntos_muestreo['Eh'].astype(float)
puntos_muestreo['Nitratos'] = puntos_muestreo['Nitratos'].astype(float)
puntos_muestreo['Sulfatos'] = puntos_muestreo['Sulfatos'].astype(int)
puntos_muestreo['Floruros'] = puntos_muestreo['Floruros'].astype(float)
puntos_muestreo['Alcalinidad'] = puntos_muestreo['Alcalinidad'].astype(int)
puntos_muestreo['Cloruros'] = puntos_muestreo['Cloruros'].astype(float)
puntos_muestreo['ICA'] = puntos_muestreo['ICA'].astype(float)

# Función para extraer valores de reflectancia en un punto
def extraer_reflectancia(archivo_banda, punto):
    with rasterio.open(archivo_banda) as src:
        x, y = punto["X"], punto["Y"]
        row, col = src.index(x, y)
        valor = src.read(1, window=((row, row+1), (col, col+1)))
        return valor[0][0]

# Función para extraer y normalizar valores RGB
def extraer_rgb_normalizado(punto):
    """
    Extrae valores RGB y los normaliza para visualización
    """
    with rasterio.open(banda_4) as src_rojo:  # Banda 4 - Rojo
        with rasterio.open(banda_3) as src_verde:  # Banda 3 - Verde
            with rasterio.open(banda_2) as src_azul:  # Banda 2 - Azul
                x, y = punto["X"], punto["Y"]

                # Obtener coordenadas de píxel para cada banda
                row_rojo, col_rojo = src_rojo.index(x, y)
                row_verde, col_verde = src_verde.index(x, y)
                row_azul, col_azul = src_azul.index(x, y)

                # Extraer valores de reflectancia
                valor_rojo = src_rojo.read(1, window=((row_rojo, row_rojo+1), (col_rojo, col_rojo+1)))[0][0]
                valor_verde = src_verde.read(1, window=((row_verde, row_verde+1), (col_verde, col_verde+1)))[0][0]
                valor_azul = src_azul.read(1, window=((row_azul, row_azul+1), (col_azul, col_azul+1)))[0][0]

                return valor_rojo, valor_verde, valor_azul

# Función para calcular RGB en formato 0-255 (para visualización)
def calcular_rgb_255(rojo, verde, azul):
    """
    Convierte valores de reflectancia a RGB 0-255 para visualización
    """
    # Normalizar los valores (puedes ajustar estos parámetros según tus datos)
    max_reflectancia = max(rojo, verde, azul, 0.3)  # 0.3 es un valor típico máximo para reflectancia

    r_255 = int((rojo / max_reflectancia) * 255)
    g_255 = int((verde / max_reflectancia) * 255)
    b_255 = int((azul / max_reflectancia) * 255)

    # Asegurar que estén en el rango 0-255
    r_255 = max(0, min(255, r_255))
    g_255 = max(0, min(255, g_255))
    b_255 = max(0, min(255, b_255))

    return r_255, g_255, b_255

# Extraer valores de reflectancia RGB para cada punto
valores_rgb = []
valores_rgb_255 = []
reflectancias = []

for idx, punto in puntos_muestreo.iterrows():
    # Extraer valores RGB
    r, g, b = extraer_rgb_normalizado(punto)
    r_255, g_255, b_255 = calcular_rgb_255(r, g, b)

    valores_rgb.append([r, g, b])
    valores_rgb_255.append([r_255, g_255, b_255])

    # Extraer otras bandas (manteniendo tu funcionalidad original)
    b1 = extraer_reflectancia(banda_1, punto)
    b5 = extraer_reflectancia(banda_5, punto)
    b6 = extraer_reflectancia(banda_6, punto)
    b7 = extraer_reflectancia(banda_7, punto)
    b8 = extraer_reflectancia(banda_8, punto)
    b8a = extraer_reflectancia(banda_8a, punto)
    b9 = extraer_reflectancia(banda_9, punto)
    b11 = extraer_reflectancia(banda_11, punto)
    b12 = extraer_reflectancia(banda_12, punto)
    ndwi = extraer_reflectancia(NDWI, punto)
    ndci = extraer_reflectancia(NDCI, punto)
    rci = extraer_reflectancia(RCI, punto)
    cdom = extraer_reflectancia(CDOM, punto)

    reflectancias.append([b1, r, g, b, b5, b6, b7, b8, b8a, b9, b11, b12, ndwi, ndci, rci, cdom])

# Imprimir resultados RGB
print("\n--- Valores RGB extraídos (primeros 5) ---")
print("Formato: [Rojo, Verde, Azul] (reflectancia)")
for i, rgb in enumerate(valores_rgb[:5]):
    print(f"Punto {i+1}: {rgb}")

print("\n--- Valores RGB en formato 0-255 (primeros 5) ---")
print("Formato: [R, G, B] (0-255 para visualización)")
for i, rgb_255 in enumerate(valores_rgb_255[:5]):
    print(f"Punto {i+1}: {rgb_255} - Color aproximado: RGB{rgb_255}")

# Agregar reflectancia de las bandas al CSV (incluyendo RGB)
df_reflectancias = pd.DataFrame(reflectancias, columns=[
    "Banda_1", "Banda_2", "Banda_3", "Banda_4",
    "Banda_5", "Banda_6", "Banda_7", "Banda_8", "Banda_8a",
    "Banda_9", "Banda_11", "Banda_12", "NDWI", "NDCI", "RCI", "CDOM"
])

# Agregar columnas RGB específicas
df_rgb = pd.DataFrame(valores_rgb, columns=["Rojo_Reflectancia", "Verde_Reflectancia", "Azul_Reflectancia"])
df_rgb_255 = pd.DataFrame(valores_rgb_255, columns=["Rojo_255", "Verde_255", "Azul_255"])

# Concatenar todos los DataFrames
puntos_muestreo_completo = pd.concat([puntos_muestreo, df_reflectancias, df_rgb, df_rgb_255], axis=1)

# Guardar el DataFrame completo
puntos_muestreo_completo.to_csv("/content/drive/MyDrive/Datos CA VT/BD Completa con RGB.csv", index=False)

# Mostrar las primeras filas
print("\n--- Primeras 5 filas del DataFrame completo ---")
print(puntos_muestreo_completo[[
    'X', 'Y', 'Rojo_Reflectancia', 'Verde_Reflectancia', 'Azul_Reflectancia',
    'Rojo_255', 'Verde_255', 'Azul_255'
]].head())

print("\n" + "="*80 + "\n")
```
